```python
import cv2 as cv
```

```python
img = cv.imread("group.jpeg")
img
```

```python
img.shape
```

```python
gray_image = cv.cvtColor(img,cv.COLOR_BGR2GRAY) # converting colorfull image to gray color image to reduce the sizee .e 2 dimensional from 3 dimensional 
gray_image
```

```python
gray_image.shape
```

```python
# creating a face classifier variable  that consist of attributes of the faces 
face_classifier = cv.CascadeClassifier(
    cv.data.haarcascades + "/haarcascade_frontalface_default.xml")
```

```python
# detecting and assigning the different attributes of faces and assigning values 
face = face_classifier.detectMultiScale( gray_image,scaleFactor = 1.1, minNeighbors = 5 , minSize = (40,40))
```

```python
for (x,y,w,h) in face:
    cv.rectangle(img,(x,y),(x+w,y+h),(0,255,0),4)
```

```python
img_rgb = cv.cvtColor(img,cv.COLOR_BGR2RGB)
```

# Access the Webcam

```python
video_capture = cv.VideoCapture(0) 
```

# this code initializes s video capture  object  using the OpenCV library in python.
* The cv.VideoCapute() functionn takes an argument that specifies the index of the camera to be used for  video.
* in this case, the argument is 0,which means the default camera on the device will be used 
* the resulting video_capute object can be used to caputre frames from the camera for further processi g or analysis .

```python
def detect_bounding_box(vid):
    gray_image = cv.cvtColor(vid,cv.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray_image,1.1,minSize =(40,40))
    for (x,y,w,h) in faces:
        cv.rectangle(vid,(x,y),(x+w,y+h),(0,255,0),4)
    return faces
```

```python
while True:
    result,video_frame = video_capture.read() # read frames from the video
    if result is False:
        break                                 # terminate the loop if the frame is not read successfully
    faces = detect_bounding_box(video_frame)                                         # apply the functionn we created to the video frame 
    cv.imshow(
    "My Face Detection project",video_frame
    ) 
    if cv.waitKey(0) & 0xFF == ord("q"):
        break

video_caputure.release
cv.destroyAllWindows()
    
```
